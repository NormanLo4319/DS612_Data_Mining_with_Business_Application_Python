```python
# Calling all the elements, but dropping the first row
A[1:]
```

```python
# Calling all the elements, but dropping the second column
A[:, [0,2,3]]
```

```python
# Calling all the elements, but dropping the second and third columns
A[:, [0,3]]
```

```python
# Random Numbers
```

```python
# Numpy has method to create random numbers
# Using rand() to create a vector x with 50 random numbers (default from 0 - 1)
x = np.random.rand(50)
print(x)
```

```python
# Creat a vector z with 50 random z-scores (Standardized scores in normal distribution)
z = np.random.standard_normal(50)
print(z)
```

```python
# Creat a vector t with 50 random t-scores (standardized scroes with mean=5 and std=10)
t = np.random.normal(5, 10, 50)
print(t)
```

```python
# Using Matplotlib to plot the data points
# Import dependency
import matplotlib.pyplot as plt

# Use scatter plot to plot the data x and z
plt.scatter(x, z)

# Give names to both axes
plt.xlabel('x - axis') 
plt.ylabel('y - axis')

# plot title 
plt.title('My scatter plot') 

# showing legend 
plt.legend() 

# Show the plot as output
plt.show()
```

```python
# Use the histogram plot to display the distribution of vector z
plt.hist(z)

plt.xlabel('z-scores') 
plt.ylabel('frequency') 

plt.show()
```

```python
# Use the histogram plot to display the distribution of vector t
plt.hist(t)

plt.xlabel('t-scores') 
plt.ylabel('frequency') 

plt.show()
```

```python
# Create a vector NN with 1,000,000 randomized standardized scores
# Creat a vector z with 50 random z-scores (Standardized scores in normal distribution)
NN = np.random.standard_normal(1000000)
```

```python
# Plot the large sample size randomized standardized scores
plt.hist(NN)
plt.xlabel('z-scores') 
plt.ylabel('frequency') 
plt.show()
```

```python
# We can also print the mean score, standard deviation, and variance of vector t
t_mean = np.mean(t)
t_std = np.std(t)
t_var = np.var(t)
print(t_mean)
print(t_std)
print(t_var)
```
